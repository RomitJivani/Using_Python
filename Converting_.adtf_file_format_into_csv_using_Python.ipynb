```python
import pandas as pd 
myfile =open("ConfigVariables.xml","r+")
```

```python
rfile =myfile.read()
rfile
```

```python
sfile = rfile.split('\n')
sfile
```

```python
for i in range(0,len(sfile)):
    temp=sfile[i]
    print(i,temp)
```

```python
str_array = []
temp_str = ""

for i in range(0,len(sfile)):
    temp=sfile[i]
    #print(i,temp)
    
    if "<ConfigVariable name=" in str(temp):
        actual_name= str(temp).split("<ConfigVariable name=")[1]
        name= str(actual_name).split(" ")[0]
        name = str(name)[1:-1]
        actual_value= str(actual_name).split("value=")[1]
        value= actual_value[1:-2]
        temp_str = str(name) + ','+ str(value)
        
    if "<id>" in str(temp):
        actual_id= str(temp).split("<id>")[1]
        id= str(actual_id).split("</id>")[0]
        temp_str = str(temp_str) +','+ str(id)
        
    if "</ConfigVariable>" in str(temp):
        str_array.append(temp_str.split(","))
        
print(str_array)
```

```python
df = pd.DataFrame(data=str_array, index=None, columns=['Name','Value','Id'])
df.to_csv('adtf_data_using_python.csv',sep =';')       #'adtf_data_using_python.txt' this name of file you will find in your folder
df
```
